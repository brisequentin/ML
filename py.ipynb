```python
import numpy as np

x = np.linspace(0, 2, 10)
y = x**2

#mathplotlib
#ne pas melanger OOP et plt.plot

import matplotlib.pyplot as plt

plt.figure(figsize = (12,8)) #init figure
plt.plot(x,y, label = 'quadratique')
plt.plot(x,y**3, label = 'cubique')
plt.title('figure 1')
plt.xlabel("axe x")
plt.ylabel("axe y")
plt.legend()
plt.show()
#plt.savefig("nomdufichier.png")

#X ET Y doivent avoir des dimensions gales

#param de custom

# c = couleur du trait
# lw = epaisseur ligne
# ls = style de ligne

#cycle de vie d'une figure
```

```python
#exo

dataset = {f"experience{i}":np.random.randn(100) for i in range(4)}

print(dataset["experience0"])

def graphique(dataset):
    n = len(dataset)
    plt.figure(figsize=(12,30)) #init figure
    for k, i in zip(dataset.keys(), range(1,n+1)):
        plt.subplot(n, 1 ,i) #ligne col pos
        plt.plot(dataset[k])
        plt.title(k)
    plt.show()

graphique(dataset)
```

```python
#pbtq classification :

import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris

iris = load_iris()

x = iris.data
y = iris.target
names = list(iris.target_names)

print(f'x contient{x.shape[0]} exemples et {x.shape[1]} variables')
print(f'il y a {np.unique(y).size} classes')

plt.scatter(x[:, 0],  x[:, 1], c=y, alpha=0.5, s=x[:,2]*100)
plt.xlabel('longueur spal')
plt.ylabel('largeur spal')
```

```python
#%matplotlib
from mpl_toolkits.mplot3d import Axes3D

ax = plt.axes(projection='3d')
ax.scatter(x[:, 0],  x[:, 1], x[:,2], c=y)

f = lambda x, y: np.sin(x) + np.cos(x+y)

X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)

X, Y = np.meshgrid(X, Y)

Z = f(X, Y)

Z.shape

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')
```

```python
#histo

#plt.hist(x[:, 0], bins=30)

plt.hist2d(x[:, 0], x[:, 1]) #cmap
plt.xlabel('longueur spal')
plt.ylabel("largeur spal")
plt.colorbar()
plt.show()


#on peut le faire aussi sur une image

from scipy import misc
face = misc.face(gray=True)
plt.imshow(face, cmap='gray')
plt.hist(face.ravel(),bins=255)
plt.show()
```
